```python
import warnings
warnings.filterwarnings('ignore')

import os
import re
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn import metrics
from autogluon.text import TextPredictor
```

데이터 load

```python
c_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/c_final_real.csv')
e_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/e_final_real.csv')
n_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/n_final_real.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/잡코리아_자소서_최종.csv')
```

train, valid 데이터 분리 함수

```python
def split_data(data):
    train_size = 0.8
    df_train = data.sample(frac=train_size, random_state=42)
    df_valid = data.drop(df_train.index).reset_index(drop=True)
    df_train.reset_index(drop=True, inplace=True)
    return df_train, df_valid
```

CON, EXT, NEU 데이터 분리

```python
train_c, valid_c = split_data(c_df)
train_e, valid_e = split_data(e_df)
train_n, valid_n = split_data(n_df)
print('CON data:', train_c.shape, valid_c.shape)
print('EXT data:', train_e.shape, valid_e.shape)
print('NEU data:', train_n.shape, valid_n.shape)
```

하이퍼파라미터 정의<br>
모델: bert-kor-base

```python
hyperparameters = {
    # https://huggingface.co/kykim/bert-kor-base
    'model.hf_text.checkpoint_name': 'kykim/bert-kor-base'
}
```

CON 데이터 학습

```python
predictorC = TextPredictor(label='cCON', problem_type='binary', eval_metric='log_loss')
predictorC.fit(train_c, tuning_data=valid_c, hyperparameters=hyperparameters)
```

CON 예측 모델 save

```python
modelC_path = '/content/drive/MyDrive/BertModel/BertModel_C'
predictorC.save(modelC_path)
# predictor = TextPredictor.load(modelC_path)
```

```python
predictorC.evaluate(valid_c, metrics=['accuracy', 'log_loss'])
```

```python
test['KOR_TEXT'] = test['A']
test['cCON'] = predictorC.predict(test[['KOR_TEXT']])
test.drop('KOR_TEXT', axis=1, inplace=True)
test.head()
```

```python
test.to_csv('/content/drive/MyDrive/Colab Notebooks/test_c.csv', index=False)
```

EXT 데이터 학습

```python
predictorE = TextPredictor(label='cEXT', problem_type='binary', eval_metric='log_loss')
predictorE.fit(train_e, tuning_data=valid_e, hyperparameters=hyperparameters)
```

EXT 예측 모델 save

```python
modelE_path = '/content/drive/MyDrive/BertModel/BertModel_E'
predictorE.save(modelE_path)
# predictor = TextPredictor.load(modelE_path)
```

```python
predictorE.evaluate(valid_e, metrics=['accuracy', 'log_loss'])
```

```python
test['KOR_TEXT'] = test['A']
test['cEXT'] = predictorE.predict(test[['KOR_TEXT']])
test.drop('KOR_TEXT', axis=1, inplace=True)
test.head()
```

```python
test.to_csv('/content/drive/MyDrive/Colab Notebooks/test_e.csv', index=False)
```

NEU 데이터 학습

```python
predictorN = TextPredictor(label='cNEU', problem_type='binary', eval_metric='log_loss')
predictorN.fit(train_n, tuning_data=valid_n, hyperparameters=hyperparameters)
```

NEU 예측 모델 save

```python
modelN_path = '/content/drive/MyDrive/BertModel/BertModel_N'
predictorN.save(modelN_path)
# predictor = TextPredictor.load(modelN_path)
```

```python
predictorN.evaluate(valid_n, metrics=['accuracy', 'log_loss'])
```

```python
test['KOR_TEXT'] = test['A']
test['cNEU'] = predictorN.predict(test[['KOR_TEXT']])
test.drop('KOR_TEXT', axis=1, inplace=True)
test.head()
```
